```python
import pandas as pd
import warnings
import numpy as np
import win32com.client as win32
warnings.filterwarnings("ignore")
```

```python
path = r"C:\Users\kfir\Downloads\אנגלית ה4.xlsx"
sheet = path.split('\\')[-1].split('.')[0]
# path.replace("\", "/", 5)
print(path)
file_name=path[:-4]

# orig_df = pd.read_csv(path)
orig_df = pd.read_excel(path, sheet_name=sheet,na_values='NaN')
```

```python
df = orig_df.iloc[3:]
```

```python
df.columns = orig_df.iloc[2]
```

```python
df['Date'] = orig_df.columns[6]
df['Class name'] = orig_df["Unnamed: 2"].loc[:0].values[0]
df['Teacher name'] = orig_df["Unnamed: 4"].loc[:0].values[0]
```

```python
df.fillna(0, inplace=True)
```

```python
df.head()
```

```python
for col in df.columns[2:-3]:
    df[col] = df[col].astype(float)
    df[col] = df[col].astype(int)
    df[col] = np.where(df[col]<40,0,df[col])
```

```python
df['# lessons'] = (df.iloc[:,2:-3]>0).sum(axis=1)
#df['number of lessons'] = (df.iloc[:,2:-3]>0).sum(axis=1)
```

```python
median = int(df['# lessons'].median())
meaidn =1 if median == 0 else median
```

```python
df_T=df.transpose().iloc[2:-4,]
df_T
```

```python
grade = []
for col in df_T.columns:
    final_grade = np.mean(df_T[col].sort_values(ascending=False).iloc[:median])
    final_grade=final_grade.astype(int)
    grade.append(final_grade)
df['Grade'] = grade
```

```python
df.head()
```

```python
df_grades = df[['Student name', 'ID number', 'Grade', '# lessons', 'Date', 'Class name', 'Teacher name']]
date = str(df['Date'].values[0])[:10].replace("/","_",2)
print(date)
#class_name = df['Class name'].values[0]
file=df_grades.to_excel(f'{file_name}_final_grades_{date}.xlsx', encoding='utf-8', freeze_panes=(1,1))
```

```python
output_file=f'{file_name}_final_grades_{date}.xlsx'
print(output_file)
```
